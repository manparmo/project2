```python
from ucimlrepo import fetch_ucirepo

from sklearn import tree
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

import pandas as pd

# fetch dataset 
taiwanese_bankruptcy_prediction = fetch_ucirepo(id=572) 

# data (as pandas dataframes) 
X = taiwanese_bankruptcy_prediction.data.features  #Rest of the DataFrame with all the features minus the Binary feature to predict
y = taiwanese_bankruptcy_prediction.data.targets   #Binary Feature to predict in the DataFrame



import project2_pipeline as pipeline
pipeline.binary_model_generator(X,y,("KNN",KNeighborsClassifier(n_neighbors=5)))
pipeline.binary_model_generator(X,y,("Logistic Regression",LogisticRegression()))
pipeline.binary_model_generator(X,y,("Random Forest",RandomForestClassifier()))
pipeline.binary_model_generator(X,y,("Decision Tree",tree.DecisionTreeClassifier()))

```
